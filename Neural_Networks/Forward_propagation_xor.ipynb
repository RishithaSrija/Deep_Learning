Neural Networks
👉 2 inputs → 2 hidden neurons → 1 output neuron

```python
import numpy as np

X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,1,1,0])
X,y
```

```python
def step_function(z):
    return 1 if z>=0 else 0
```

```python
w1,w2,w3,w4,w5,w6=1,1,1,1,1,-2
b1,b2,b3=-0.5,-1.5,-0.5

```

```python
for x1,x2 in X:
    
    h1=w1*x1+w3*x2+b1
    h2=w2*x1+w4*x2+b2

    a_h1=step_function(h1)
    a_h2=step_function(h2)

    y=w5*a_h1+w6*a_h2+b3
    a_y=step_function(y)
    print(f"Input: ({x1},{x2}) → Output: {a_y}")
```
