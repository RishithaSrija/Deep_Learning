Diabetes Risk Prediction (Binary Classification)

2 inputs -> 3Hidden Neurons(Sigmoid) -> Output(Sigmoid)

output= 1 -> High risk / = 0 ->Low risk

```python
import numpy as np
np.random.seed(0)

X=np.array([
    [120,85,150,95],    #Glocuse
    [25,30,35,22]       #BMI
])

X=X/200
Y=np.array([[1,0,1,0]])

print("X shape: ",X.shape)
print("Y shape: ",Y.shape)
```

Sigmoid Function

```python
def sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

Initialize Parameters

```python
#Hidden Layer(3 neurons,2 inputs)
w1=np.random.randn(3,2)*0.1
b1=np.zeros((3,1))

#Output layer
w2=np.random.randn(1,3)*0.1
b2=np.zeros((1,1))
```

Forward Propagation

```python
#Hidden Layer
z1=np.dot(w1,X)+b1
a1=sigmoid(z1)

#Output Layer
z2=np.dot(w2,a1)+b2
a2=sigmoid(z2)

print("Predicted Output: ")
print(a2)
```

Loss Function

Binary Cross Entropy

```python
m=Y.shape[1]
loss=-np.sum(Y*np.log(a2+1e-8)+(1-Y)*np.log(1-a2+1e-8))/m 
print(" Loss : ",loss)
```

Back Propagation

Output Layer Gradient

```python
dz2=a2-Y
dw2=(1/m)*np.dot(dz2,a1.T)
db2=(1/m)*np.sum(dz2,axis=1,keepdims=True)
```

Hidden Layer Gradient

Sigmoid derivative:

```python
dA1=np.dot(w2.T,dz2)
dz1=dA1*a1*(1-a1)

dw1=(1/m)*np.dot(dz1,X.T)
db1=(1/m)*np.sum(dz1,axis=1,keepdims=True)
```

Update Parameters

```python
learning_rate=0.5
w2-=learning_rate*dw2
b2=learning_rate*db2

w1-=learning_rate*dw1
b1-=learning_rate*db1
```

Complete Training Loop

```python
epoch=5000
for i in range(epoch):
    #forward
    Z1=np.dot(w1,X)+b1
    A1=sigmoid(Z1)

    Z2=np.dot(w2,A1)+b2
    A2=sigmoid(Z2)

    #loss
    loss=-np.sum(Y*np.log(A2+1e-8)+(1-Y)*np.log(1-A2+1e-8))/m 

    #backward
    dz2=A2-Y
    dw2=(1/m)*np.dot(dz2,A1.T)
    db2=(1/m)*np.sum(dz2,axis=1,keepdims=True)

    dA1=np.dot(w2.T,dz2)
    dz1=dA1*A1*(1-A1)

    dw1=(1/m)*np.dot(dz1,X.T)
    db1=(1/m)*np.sum(dz1,axis=1,keepdims=True)

    #Update
    w2-=0.5*dw2
    b2-=0.5*db2
    w1-=0.5*dw1
    b1-=0.5*db1

    if i%100==0:
        print("Epoch : ",i,"Loss: ",loss)

```
