Perceptron(Binary Classification)

```python
import numpy as np
```

Step-1: Prepare Training Data

```python
X=np.array([1,2,3,4])
y=np.array([0,0,1,1])
X,y
```

Initialize the params

```python
w=0.0
b=0.0
learning_rate=1
epochs=5
print("Initial Weight: ",w)
print("Initial Bias: ",b)
```

Defining Activation function

```python
def step_function(z):
    return 1 if z>=0 else 0
```

*** Training Loop ***

```python
import matplotlib.pyplot as plt

weight_history = []
error_history = []

for epoch in range(epochs):
    print(f'\nEpoch {epoch+1}')
    for i in range(len(X)):
        #1.Compute weighted sum
        z=w*X[i]+b

        #2.Apply Activation Function
        y_pred=step_function(z)

        #3.Compute Error
        error=y[i]-y_pred

        #4.Update Weight
        w=w+learning_rate* error *X[i]

        #5.Update bias
        b=b+learning_rate*error

        weight_history.append(w)
        error_history.append(error)
        
        print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, w: {w}, b: {b}")

plt.figure()
plt.plot(weight_history)
plt.title("Weight Updates During Training")
plt.xlabel("Step")
plt.ylabel("Weight")
plt.show()

# Plot error changes
plt.figure()
plt.plot(error_history)
plt.title("Error During Training")
plt.xlabel("Step")
plt.ylabel("Error")
plt.show()
```

1️⃣ Model is already perfect (this is OK)

If your error is always 0:

👉 Your perceptron is predicting correctly from the beginning.

```python
print("Final  Weight: ",w)
print("Final Bias: ",b)
```

Testing the Model

```python
test_value=2.9
z=w*test_value+b
prediction=step_function(z)

print(f"For study hours: {test_value}, Prediction: ",prediction)
```
