*** Problem: Smart Loan Approval System ***

A fintech company wants to automate loan approval using AI.

🔹 Initial Setup (2 Inputs)

Each applicant has:

x₁ → Income Level

0 = Low

1 = High

x₂ → Credit Score Category

0 = Poor

1 = Good

🔹 Original Approval Rules

Domain experts define:

If income is high AND credit score is good → Approve

If income is low AND credit score is poor → Reject

If one is high and the other is low → Further investigation → Approve only if credit score is good

🔹 Resulting Decision Table
Income (x₁)	Credit (x₂)	Approve
0	0	0
0	1	1
1	0	0
1	1	1

Inputs:
x1 = income
x2 = credit score
x3 = existing loan

Rule:

Approve = (x2 AND NOT x3) OR (x1 AND x2)

🔹 Policy Upgrade

New policy states:

Approve if:

Income is high AND credit is good
OR

Income is low AND credit is good

Reject if:

Both are poor
OR

Income high but credit poor

🔹 New Truth Table (Still 2 Inputs)
Income	Credit	Approve
0	0	0
0	1	1
1	0	0
1	1	1

Again simplifies to:

Approve = Credit

So still dependent only on credit score.

```python
def step_function(z):
    return 1 if z>=0 else 0
```



🔹 Further Modification (Add Risk Feature)

Now introduce a third input:

x₃ → Existing Loan

0 = No

1 = Yes

🔹 New Complex Rule

Approve if:

Credit is good AND no existing loan
OR

Income high AND credit good

Reject otherwise.

🔹 Final 3-Input Truth Table
x₁ (Income)	x₂ (Credit)	x₃ (Loan)	Approve

0	1	0	1
